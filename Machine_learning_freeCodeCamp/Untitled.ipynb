```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# defining the name of the columns manually because they are not given in the data
cols = ['fLength', 'fWidth', 'fSize', 'fConc', 'fConc1','fAsym','fM3Long', 'fM3Trans', 'fAlpha', 'fDist', 'class']
df = pd.read_csv('magic04.data', names=cols)
```

```python
df
```

```python
# converting the label into binary 
df['class'] = (df['class'] == 'g').astype(int)
```

```python
df['class'].unique() # 1 is for 'g' and 0 is for 'h'
```

```python
df.head()
```

```python
# now we are going to check the values of the features for each class.
# for gamma
df_gamma = df[df['class'] == 1]
df_gamma
```

```python
# now to check for hadron
df_hadron = df[df['class'] == 0]
df_hadron
```

```python
for col in df[:10]:
    plt.hist([df_gamma[col],df_hadron[col]], color = ['blue', 'red'], bins = 'auto', label = ['Gamma', 'Hadron'])
#     plt.hist(df_hadron[col], color = 'red', bins = 'auto', label = 'Hadron')
    plt.legend()
    plt.title(col)
    plt.show()
```
